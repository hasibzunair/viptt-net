```python
import os
import numpy as np
import pickle
import tensorflow as tf
from tqdm import tqdm
from tensorflow.keras.models import load_model


ROOT_DIR = os.path.abspath("../")
LOG_PATH = os.path.join(ROOT_DIR, "logs")
print(LOG_PATH)
```

```python
f1 = open("../logs/augmented_submission.txt", "r")
f2 = open("../logs/video_pretraining_all_layer_aug_cw_submission.txt", "r")
f3 = open("../logs/video_pretraining_all_layer_cw_submission.txt", "r")


p1 = []
p2 = []
p3 = []
names = []

for x in f1:
    p1.append(int(x[16:17]))
    names.append(x[:15])
    
    
for x in f2:
    p2.append(int(x[16:17]))
    
for x in f3:
    p3.append(int(x[16:17]))
    
    
len(p1), len(p2), len(p3), len(names)
```

```python
p1 = np.array(p1)
p2 = np.array(p2)
p3 = np.array(p3)
```

```python
p1[0], p2[0], p3[0]
```

```python
# ensemble
pred = .6 *p1 + .2 *p2 + .2 *p3 
pred.shape
```

```python
pred = [int(round(x)) for x in pred]
len(pred)
```

```python
with open('{}/{}_submission.txt'.format(LOG_PATH, "ensemble"), 'w') as f:
    for n, p in zip(names, pred):
        print(n,",", p)
        f.write(str(n))
        f.write(",")
        f.write(str(p))
        f.write("\n")
        
f.close()
```
